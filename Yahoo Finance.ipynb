```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
#resources
btc_csv= "Resources/BTC-USD 5y monthly.csv"
eth_csv= "Resources/ETH-USD 5y monthly.csv"
ltc_csv= "Resources/LTC-USD 5y monthly.csv"
bch_csv= "Resources/BNB-USD 5y monthly.csv"
bnb_csv= "Resources/BCH-USD 5y monthly.csv"
```

```python
#Read Bitcoin(BTC)
btc_df=pd.read_csv(btc_csv)
btc_renamed=btc_df.rename(columns={"Open": "BTC Open in USD",
                                "High": "BTC High in USD",
                                "Low": "BTC Low in USD",
                                "Close": "BTC Close in USD",
                                "Adj Close": "BTC Adj Close in USD",
                                "Volume": "BTC Volume in Billions"})
btc_renamed["BTC Open in USD"] = btc_renamed["BTC Open in USD"].map("{:.2f}".format)
btc_renamed["BTC High in USD"] = btc_renamed["BTC High in USD"].map("{:.2f}".format)
btc_renamed["BTC Low in USD"] = btc_renamed["BTC Low in USD"].map("{:.2f}".format)
btc_renamed["BTC Close in USD"] = btc_renamed["BTC Close in USD"].map("{:.2f}".format)
btc_renamed["BTC Adj Close in USD"] = btc_renamed["BTC Adj Close in USD"].map("{:.2f}".format)
btc_renamed["BTC Volume in Billions"]=btc_renamed["BTC Volume in Billions"]/1000000000
btc_renamed["BTC Volume in Billions"] = btc_renamed["BTC Volume in Billions"].map("{:.2f}".format)
btc_renamed.head()
```

```python
#Read Ether(ETH)
eth_df=pd.read_csv(eth_csv)
eth_renamed=eth_df.rename(columns={"Open": "ETH Open in USD",
                                "High": "ETH High in USD",
                                "Low": "ETH Low in USD",
                                "Close": "ETH Close in USD",
                                "Adj Close": "ETH Adj Close in USD",
                                "Volume": "ETH Volume in Billions"})
eth_renamed["ETH Open in USD"] = eth_renamed["ETH Open in USD"].map("{:.2f}".format)
eth_renamed["ETH High in USD"] = eth_renamed["ETH High in USD"].map("{:.2f}".format)
eth_renamed["ETH Low in USD"] = eth_renamed["ETH Low in USD"].map("{:.2f}".format)
eth_renamed["ETH Close in USD"] = eth_renamed["ETH Close in USD"].map("{:.2f}".format)
eth_renamed["ETH Adj Close in USD"] = eth_renamed["ETH Adj Close in USD"].map("{:.2f}".format)
eth_renamed["ETH Volume in Billions"]=eth_renamed["ETH Volume in Billions"]/1000000000
eth_renamed["ETH Volume in Billions"] = eth_renamed["ETH Volume in Billions"].map("{:.2f}".format)
eth_renamed.head()
```

```python
#Read Litecoin(LTC)
ltc_df=pd.read_csv(ltc_csv)
ltc_renamed=ltc_df.rename(columns={"Open": "LTC Open in USD",
                                "High": "LTC High in USD",
                                "Low": "LTC Low in USD",
                                "Close": "LTC Close in USD",
                                "Adj Close": "LTC Adj Close in USD",
                                "Volume": "LTC Volume in Billions"})
ltc_renamed["LTC Open in USD"] = ltc_renamed["LTC Open in USD"].map("{:.2f}".format)
ltc_renamed["LTC High in USD"] = ltc_renamed["LTC High in USD"].map("{:.2f}".format)
ltc_renamed["LTC Low in USD"] = ltc_renamed["LTC Low in USD"].map("{:.2f}".format)
ltc_renamed["LTC Close in USD"] = ltc_renamed["LTC Close in USD"].map("{:.2f}".format)
ltc_renamed["LTC Adj Close in USD"] = ltc_renamed["LTC Adj Close in USD"].map("{:.2f}".format)
ltc_renamed["LTC Volume in Billions"]=ltc_renamed["LTC Volume in Billions"]/1000000000
ltc_renamed["LTC Volume in Billions"] = ltc_renamed["LTC Volume in Billions"].map("{:.2f}".format)
ltc_renamed.head()
```

```python
#Read BitcoinCash(BCH)
bch_df=pd.read_csv(bch_csv)
bch_renamed=bch_df.rename(columns={"Open": "BCH Open in USD",
                                "High": "BCH High in USD",
                                "Low": "BCH Low in USD",
                                "Close": "BCH Close in USD",
                                "Adj Close": "BCH Adj Close in USD",
                                "Volume": "BCH Volume in Billions"})
bch_renamed["BCH Open in USD"] = bch_renamed["BCH Open in USD"].map("{:.2f}".format)
bch_renamed["BCH High in USD"] = bch_renamed["BCH High in USD"].map("{:.2f}".format)
bch_renamed["BCH Low in USD"] = bch_renamed["BCH Low in USD"].map("{:.2f}".format)
bch_renamed["BCH Close in USD"] = bch_renamed["BCH Close in USD"].map("{:.2f}".format)
bch_renamed["BCH Adj Close in USD"] = bch_renamed["BCH Adj Close in USD"].map("{:.2f}".format)
bch_renamed["BCH Volume in Billions"]=bch_renamed["BCH Volume in Billions"]/1000000000
bch_renamed["BCH Volume in Billions"] = bch_renamed["BCH Volume in Billions"].map("{:.2f}".format)
bch_renamed.head()
```

```python
#Read BinanceCoin(BNB)
bnb_df=pd.read_csv(bnb_csv)
bnb_renamed=bnb_df.rename(columns={"Open": "BNB Open in USD",
                                "High": "BNB High in USD",
                                "Low": "BNB Low in USD",
                                "Close": "BNB Close in USD",
                                "Adj Close": "BNB Adj Close in USD",
                                "Volume": "BNB Volume in Billions"})
bnb_renamed["BNB Open in USD"] = bnb_renamed["BNB Open in USD"].map("{:.2f}".format)
bnb_renamed["BNB High in USD"] = bnb_renamed["BNB High in USD"].map("{:.2f}".format)
bnb_renamed["BNB Low in USD"] = bnb_renamed["BNB Low in USD"].map("{:.2f}".format)
bnb_renamed["BNB Close in USD"] = bnb_renamed["BNB Close in USD"].map("{:.2f}".format)
bnb_renamed["BNB Adj Close in USD"] = bnb_renamed["BNB Adj Close in USD"].map("{:.2f}".format)
bnb_renamed["BNB Volume in Billions"]=bnb_renamed["BNB Volume in Billions"]/1000000000
bnb_renamed["BNB Volume in Billions"] = bnb_renamed["BNB Volume in Billions"].map("{:.2f}".format)
bnb_renamed.head()
```

```python
#Merge btc_eth
btc_eth=pd.merge(btc_renamed, eth_renamed, on="Date", how="outer")
#Merge ltc_bch
ltc_bch=pd.merge(ltc_renamed, bch_renamed, on="Date", how="outer")
#Merge ltc_bch with BNB
ltc_bch_bnb=pd.merge(ltc_bch, bnb_renamed, on="Date", how="outer")
#Merge All
crypto_df=pd.merge(btc_eth, ltc_bch_bnb, on="Date", how="outer")
#Replace Nan with 0
crypto_df["BCH Open in USD"]=crypto_df["BCH Open in USD"].fillna(0)
crypto_df["BCH High in USD"]=crypto_df["BCH High in USD"].fillna(0)
crypto_df["BCH Low in USD"]=crypto_df["BCH Low in USD"].fillna(0)
crypto_df["BCH Close in USD"]=crypto_df["BCH Close in USD"].fillna(0)
crypto_df["BCH Adj Close in USD"]=crypto_df["BCH Adj Close in USD"].fillna(0)
crypto_df["BCH Volume in Billions"]=crypto_df["BCH Volume in Billions"].fillna(0)
crypto_df["BNB Open in USD"]=crypto_df["BNB Open in USD"].fillna(0)
crypto_df["BNB High in USD"]=crypto_df["BNB High in USD"].fillna(0)
crypto_df["BNB Low in USD"]=crypto_df["BNB Low in USD"].fillna(0)
crypto_df["BNB Close in USD"]=crypto_df["BNB Close in USD"].fillna(0)
crypto_df["BNB Adj Close in USD"]=crypto_df["BNB Adj Close in USD"].fillna(0)
crypto_df["BNB Volume in Billions"]=crypto_df["BNB Volume in Billions"].fillna(0)
#Dropping todays date from dataframe
crypto_df.drop(crypto_df.tail(-60).index,inplace=True)
#DataFrame
crypto_df
#Write CSV to verify data is all there
#crypto_df.to_csv("Output/Crypto.csv", index=False, header=True)
```
