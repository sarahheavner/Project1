```python
import pandas as pd
import matplotlib.pyplot as plt

file = '../../Resources/Crypto.csv'

crypto_df = pd.read_csv(file)

crypto_df
```

```python
volume_df = crypto_df[['Date','BTC Volume in Billions','ETH Volume in Billions','LTC Volume in Billions','BCH Volume in Billions','BNB Volume in Billions']].set_index('Date')
volume_df
```

```python
volume_df.plot.line(subplots=True, figsize=(10,10))
plt.show()
```

```python
close_df = crypto_df[['Date','BTC Close in USD','ETH Close in USD','LTC Close in USD','BCH Close in USD','BNB Close in USD']].set_index('Date')

close_df.plot.line(subplots=True, figsize=(10,10))
plt.show()
```
