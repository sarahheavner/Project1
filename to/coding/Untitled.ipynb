```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'Resources/Crypto.csv'

crypto_df = pd.read_csv(file)

crypto_df
```

```python
volume_df = crypto_df[['Date','BTC Volume in Billions','ETH Volume in Billions','LTC Volume in Billions','BCH Volume in Billions','BNB Volume in Billions']].set_index('Date')
volume_df
```

```python
volume_df.plot.line(subplots=True, figsize=(10,10))
plt.show()
```
