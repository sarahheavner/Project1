```python
import requests
import pandas as pd
from pprint import pprint

crypto_list = ['BTC', 'ETH', 'LTC', 'BCH', 'BNB']
crypto_id = []

for coin in crypto_list:

    url = "https://bravenewcoin.p.rapidapi.com/asset"

    querystring = {"status":"ACTIVE","symbol": coin}

    headers = {
        'x-rapidapi-host': "bravenewcoin.p.rapidapi.com",
        'x-rapidapi-key': "0c17ec419bmsh21887ba3f0668d1p1885e5jsneab3fc991989"
        }

    response = requests.request("GET", url, headers=headers, params=querystring).json()
    
    crypto_id.append(response['content'][0]['id'])
```

```python
pprint(response)
```

```python
url = "https://bravenewcoin.p.rapidapi.com/oauth/token"

payload = '{ "audience": "https://api.bravenewcoin.com", "client_id": "oCdQoZoI96ERE9HY3sQ7JmbACfBf55RY", "grant_type": "client_credentials"}'
headers = {
    'x-rapidapi-host': "bravenewcoin.p.rapidapi.com",
    'x-rapidapi-key': "0c17ec419bmsh21887ba3f0668d1p1885e5jsneab3fc991989",
    'content-type': "application/json",
    'accept': "application/json"
    }

token = requests.request("POST", url, data=payload, headers=headers).json()

access_token = token['access_token']
```

```python
ticker_list = []

for coin in crypto_id:
    
    url = "https://bravenewcoin.p.rapidapi.com/market-cap"

    querystring = {"assetId": coin}

    headers = {
        'x-rapidapi-host': "bravenewcoin.p.rapidapi.com",
        'x-rapidapi-key': "0c17ec419bmsh21887ba3f0668d1p1885e5jsneab3fc991989",
        'authorization': f"Bearer {access_token}"
        }

    ticker = requests.request("GET", url, headers=headers, params=querystring).json()
    
    ticker_list.append(ticker['content'][0])
```

```python
bnc_df = pd.DataFrame(ticker_list)
bnc_df['Ticker'] = crypto_list
bnc_df.set_index('Ticker')
```
