```python
#Dependencies
import glob
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
#Glob data to combine all CSV's
path= r'C:\Users\Phil\Desktop\Project\Resources'
all_files=glob.glob(path+"/*.csv")
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
#Column clean up and formatting
frame=frame.rename(columns={"Open": "Open in USD",
                                "High": "High in USD",
                                "Low": "Low in USD",
                                "Close": "Close in USD",
                                "Adj Close": "Adj Close in USD",
                                "Volume": "Volume in Billions"})
frame["Open in USD"] = frame["Open in USD"].map("{:.2f}".format)
frame["High in USD"] = frame["High in USD"].map("{:.2f}".format)
frame["Low in USD"] = frame["Low in USD"].map("{:.2f}".format)
frame["Close in USD"] = frame["Close in USD"].map("{:.2f}".format)
frame["Adj Close in USD"] = frame["Adj Close in USD"].map("{:.2f}".format)
frame["Volume in Billions"]=frame["Volume in Billions"]/1000000000
frame["Volume in Billions"] = frame["Volume in Billions"].map("{:.2f}".format)
#Write CSV
#frame.to_csv("Output/CryptoCurrency.csv", index=False, header=True)
frame["Open in USD"]=frame["Open in USD"].astype(str).astype(float)
frame["High in USD"]=frame["High in USD"].astype(str).astype(float)
frame["Low in USD"]=frame["Low in USD"].astype(str).astype(float)
frame["Close in USD"]=frame["Close in USD"].astype(str).astype(float)
frame["Adj Close in USD"]=frame["Adj Close in USD"].astype(str).astype(float)
frame["Volume in Billions"]=frame["Volume in Billions"].astype(str).astype(float)
frame.dtypes
frame.head()
```
